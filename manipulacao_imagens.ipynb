```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Carregamento das imagens e união de todas elas em uma única imagem

```python
imagemA0 = cv2.imread("A\A0.jpeg")
imagemA1 = cv2.imread("A\A1.jpeg")
imagemA2 = cv2.imread("A\A2.jpeg")
imagemA3 = cv2.imread("A\A3.jpeg")
imagemA4 = cv2.imread("A\A4.jpeg")
imagemA5 = cv2.imread("A\A5.jpeg")
imagemA6 = cv2.imread("A\A6.jpeg")
imagemA7 = cv2.imread("A\A7.jpeg")
imagemA8 = cv2.imread("A\A8.jpeg")
imagemA9 = cv2.imread("A\A9.jpeg")

A1 = np.concatenate((imagemA0, imagemA1, imagemA2, imagemA3, imagemA4), axis=1)
A2 = np.concatenate((imagemA5, imagemA6, imagemA7, imagemA8, imagemA9), axis = 1)
A = np.concatenate((A1, A2), axis = 0)

A_pb = cv2.cvtColor(A, cv2.COLOR_BGR2GRAY)
print(A_pb.shape) #Resolução x 1
plt.imshow(A_pb, cmap='gray')
plt.show()
```

```python
imagemB0 = cv2.imread("B\B0.jpeg")
imagemB1 = cv2.imread("B\B1.jpeg")
imagemB2 = cv2.imread("B\B2.jpeg")
imagemB3 = cv2.imread("B\B3.jpeg")
imagemB4 = cv2.imread("B\B4.jpeg")
imagemB5 = cv2.imread("B\B5.jpeg")
imagemB6 = cv2.imread("B\B6.jpeg")
imagemB7 = cv2.imread("B\B7.jpeg")
imagemB8 = cv2.imread("B\B8.jpeg")
imagemB9 = cv2.imread("B\B9.jpeg")

B1 = np.concatenate((imagemB0, imagemB1, imagemB2, imagemB3, imagemB4), axis=1)
B2 = np.concatenate((imagemB5, imagemB6, imagemB7, imagemB8, imagemB9), axis = 1)
B = np.concatenate((B1, B2), axis = 0)

B_pb = cv2.cvtColor(B, cv2.COLOR_BGR2GRAY)
plt.imshow(B_pb, cmap='gray')
plt.show()
```

```python
imagemC0 = cv2.imread("C\C0.jpeg")
imagemC1 = cv2.imread("C\C1.jpeg")
imagemC2 = cv2.imread("C\C2.jpeg")
imagemC3 = cv2.imread("C\C3.jpeg")
imagemC4 = cv2.imread("C\C4.jpeg")
imagemC5 = cv2.imread("C\C5.jpeg")
imagemC6 = cv2.imread("C\C6.jpeg")
imagemC7 = cv2.imread("C\C7.jpeg")
imagemC8 = cv2.imread("C\C8.jpeg")
imagemC9 = cv2.imread("C\C9.jpeg")

C1 = np.concatenate((imagemC0, imagemC1, imagemC2, imagemC3, imagemC4), axis=1)
C2 = np.concatenate((imagemC5, imagemC6, imagemC7, imagemC8, imagemC9), axis = 1)
C = np.concatenate((C1, C2), axis = 0)

C_pb = cv2.cvtColor(C, cv2.COLOR_BGR2GRAY)
plt.imshow(C_pb, cmap='gray')
plt.show()
```

```python
ABC_pb = np.concatenate((A_pb, B_pb, C_pb), axis = 1)
plt.imshow(ABC_pb, cmap = "gray")
plt.show()
```

## Salvando a imagem para aplicação do algoritmo de machine learning
